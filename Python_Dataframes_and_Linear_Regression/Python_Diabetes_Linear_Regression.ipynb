```python
%matplotlib notebook
import pandas as pd
import numpy as np
from sklearn import datasets,linear_model
import matplotlib.pyplot as plt
```

```python
Data = datasets.load_diabetes()
```

```python
Diabetes = pd.DataFrame(Data.data,columns = Data.feature_names)
Diabetes['target'] = pd.Series(Data.target)
Diabetes.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(Diabetes.iloc[:,:10],Diabetes['target'],random_state =6)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train,Y_train)
regr.score(X_test,Y_test)
```

```python
print(f"coeffient:{regr.coef_}")
print(f"intercept:{regr.intercept_}")
```

```python
Y_predict = regr.predict(X_test)
```

```python
plt.scatter(Y_test,Y_predict, color='blue')
# y=x line for reference
plt.plot( Y_test , Y_test ,linewidth = 3 ,color='black')
plt.show()
```
